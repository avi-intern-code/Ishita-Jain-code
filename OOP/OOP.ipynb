```python
class comp:
        
        def __init__(self):
            self.name='Naveen'
            self.age=20

        def compare(self, other):
            if self.age==other.age :
                return True
            else:
                return False
            
c1=comp()
c2=comp()

c1.name='Rashi'
c1.age=12

if c1.compare(c2):
    print('They are same')
else:
    print('They are not same')


print(c1.name, c1.age)
print(c2.name, c2.age)
```

```python
class student:
    university = 'JKLU'
    
    def __init__(self,m1,m2,m3):
        self.m1=m1
        self.m2=m2
        self.m3=m3
        
    def avg(self):
        return (self.m1+self.m2+self.m3)/3
    
    @classmethod
    def info(cls):
        return cls.university
    
    @staticmethod
    def year():
        print('2nd year')

s1=student(12,34,47)
s2=student(63,87,21)

print(student.info())
student.year()
print(s1.avg())
print(s2.avg())
```

```python
class student:
    def __init__ (self, name, rollno):
        self.name=name
        self.rollno=rollno
        self.lap=self.laptop()
        
    def show(self):
        print(self.name,self.rollno)
        self.lap.show()
        
    class laptop:
        def __init__(self):
            self.brand='HP'
            self.cpu='i5'
            self.ram=8
        
        def show(self):
            print(self.brand, self.cpu, self.ram)

s1=student('Naveen',2)
s2=student('Rashi',3)

s1.show()
s2.show()
```

```python
#inheritance

class a:
    def f1(self):
        print('f1 is working')
        
    def f2(self):
        print('f2 is working')
    
class b:
    def f3(self):
        print('f3 is working')
        
    def f4(self):
        print('f4 is working')

class c(a,b):
     def f5(self):
        print('f5 is working')
            
     def f6(self):
        print('f6 is working')
        
a1=a()
b1=b()
c1=c()

a1.f1()
b1.f3()
c1.f2()
```

```python
class a:
    
    def __init__(self):
        print('in a')
        
    def f1(self):
        print('f1 is working')
        
    def f2(self):
        print('f2 is working')
    
class b(a):
    
    def __init__(self):
        super().__init__()
        print('in b')
        
    def f3(self):
        print('f3 is working')
        
    def f4(self):
        print('f4 is working')
        
a1=b()
```

```python
class a:
    def __init__(self,m1,m2):
        self.m1=m1
        self.m2=m2
        
    def __add__(self,other):
        m1=self.m1+other.m1
        m2=self.m2+other.m2
        s3=a(m1,m2)
        
        return s3
    
    def __gt__(self,other):
        x=self.m1+self.m2
        y=other.m1+other.m2
        if x>y:
            return True
        else:
            return False
    def __str__(self):
        return self.m1, self.m2
    
s1=a(12,35)
s2=a(64,92)

print(s1.__str__())
print(s2.__str__())

s3=s1+s2
print(s3.m1,s3.m2)

if s1>s2:
            print('s1 wins')
else:
            print('s2 wins')
```
